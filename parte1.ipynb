```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

## Grafos Modelo _ERDÖS RÉNYE_:


```python

ER_1 = nx.erdos_renyi_graph(n = 100, p = 0.05, seed = 42)
grau_1 = [grau_1 for no, grau_1 in ER_1.degree()] #quantidade de nós de cada grafo

ER_2 = nx.erdos_renyi_graph(n = 100, p = 0.02, seed = 42)
grau_2 = [grau_2 for no, grau_2 in ER_2.degree()]

ER_3 = nx.erdos_renyi_graph(n = 100, p = 0.10, seed = 42)
grau_3 = [grau_3 for no, grau_3 in ER_3.degree()]
```

### **Primeiro Grafo (Erdös-Rényi)**:

```python
nx.draw(ER_1);
plt.show
```

```python
bins = np.arange(min(grau_1), max(grau_1) + 2) - 0.5

plt.figure(figsize=(7, 5))
plt.hist(grau_1, bins=bins, color="darkorange", edgecolor="black")
plt.xlabel("Grau (k)")
plt.ylabel("Frequência (número de nós)")
plt.title("Primeira Distribuição de graus - Erdős-Rényi")
plt.xticks(range(min(grau_1), max(grau_1)+1))
plt.show()
```

### Dados do Grafo:

```python
gm_er1 = np.mean(grau_1)
print ("grau médio:", gm_er1)

desvpad_er1 = np.std(grau_1) 
print ("\ndesvio padrão:", desvpad_er1)

densidade_er1 = nx.density(ER_1) 
print ("\ndensidade:", densidade_er1)

midway_er1 = nx.average_shortest_path_length(ER_1)
print ("\ncaminho médio:", midway_er1)

diameter_er1 = nx.diameter(ER_1)
print("\ndiâmetro:", diameter_er1)

midcluster_er1 = nx.average_clustering(ER_1)
print ("\ncoeficiênte de clustering médio:", midcluster_er1)
```
