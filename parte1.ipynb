```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

## Grafos Modelo _ERDÖS RÉNYE_:


```python

ER_1 = nx.erdos_renyi_graph(n = 100, p = 0.05, seed = 42) # rede próxima do limiar
grau_er1 = [grau_er1 for no, grau_er1 in ER_1.degree()] 

ER_2 = nx.erdos_renyi_graph(n = 100, p = 0.02, seed = 42) # rede esparsa
grau_er2 = [grau_er2 for no, grau_er2 in ER_2.degree()]

ER_3 = nx.erdos_renyi_graph(n = 100, p = 0.10, seed = 42) # rede densa
grau_er3 = [grau_er3 for no, grau_er3 in ER_3.degree()]
```

### **Primeiro Grafo (Erdös-Rényi)**
### _limiar próximo_:

```python
nx.draw(ER_1);
plt.show
```

```python
bins = np.arange(min(grau_er1), max(grau_er1) + 2) - 0.5

plt.figure(figsize=(7, 5))
plt.hist(grau_er1, bins=bins, color="darkorange", edgecolor="black")
plt.xlabel("Grau (k)")
plt.ylabel("Frequência (número de nós)")
plt.title("Primeira Distribuição de graus - Erdős-Rényi")
plt.xticks(range(min(grau_er1), max(grau_er1)+1))
plt.show()
```

### Dados do Grafo:

```python
gm_er1 = np.mean(grau_er1)
print ("grau médio:", gm_er1)

desvpad_er1 = np.std(grau_er1) 
print ("\ndesvio padrão:", desvpad_er1)

densidade_er1 = nx.density(ER_1) 
print ("\ndensidade:", densidade_er1)

midway_er1 = nx.average_shortest_path_length(ER_1)
print ("\ncaminho médio:", midway_er1)

diameter_er1 = nx.diameter(ER_1)
print("\ndiâmetro:", diameter_er1)

midcluster_er1 = nx.average_clustering(ER_1)
print ("\ncoeficiênte de clustering médio:", midcluster_er1)
```

## Grafos Modelo _Watts-Strogatz_:

```python
WS_1 = nx.watts_strogatz_graph(n = 100, k = 6, p = 0.0, seed = 42) # anel puro
grau_ws1 = [grau_er1 for no, grau_er1 in ER_1.degree()]

WS_2 = nx.watts_strogatz_graph(n = 100, k = 6, p = 0.05, seed = 42) # mundo pequeno
grau_ws2 = [grau_er1 for no, grau_er1 in ER_1.degree()]

WS_3 = nx.watts_strogatz_graph(n = 100, k = 6, p = 0.5, seed = 42) # em transição
grau_ws3 = [grau_er1 for no, grau_er1 in ER_1.degree()]
```
