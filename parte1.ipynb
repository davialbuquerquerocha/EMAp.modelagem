```python
def calcular_componentes_e_centralidades(G, nome="Rede"):
    componentes = list(nx.connected_components(G))
    n_componentes = len(componentes)
    maior_componente = max(componentes, key=len)
    tamanho_maior = len(maior_componente)

    grau_cent = nx.degree_centrality(G)
    prox_cent = nx.closeness_centrality(G)
    inter_cent = nx.betweenness_centrality(G)
    top_hub = max(grau_cent.items(), key=lambda x: x[1])

    print(f"--- Componentes e hubs ({nome}) ---")
    print("Número de componentes conectados:", n_componentes)
    print("Tamanho do maior componente:", tamanho_maior)
    print(f"Nó com maior centralidade de grau: {top_hub[0]} (centralidade={top_hub[1]:.4f})")
    print(f"Centralidade de proximidade desse nó: {prox_cent[top_hub[0]]:.4f}")
    print(f"Centralidade de intermediação desse nó: {inter_cent[top_hub[0]]:.4f}")

    return n_componentes, tamanho_maior, grau_cent, prox_cent, inter_cent
```

```python
def analise_transicao_ws(n, k, valores_p, seed=42, n_repeticoes=5):
    G0 = nx.watts_strogatz_graph(n, k, 0, seed=seed)
    C0 = nx.average_clustering(G0)
    L0 = nx.average_shortest_path_length(G0)

    resultados_p, resultados_C, resultados_L = [], [], []
    for p in valores_p:
        Cs, Ls = [], []
        for r in range(n_repeticoes):
            G = nx.watts_strogatz_graph(n, k, p, seed=seed + r)
            if nx.is_connected(G):
                L = nx.average_shortest_path_length(G)
            else:
                maior = max(nx.connected_components(G), key=len)
                L = nx.average_shortest_path_length(G.subgraph(maior))
            Cs.append(nx.average_clustering(G))
            Ls.append(L)
        resultados_p.append(p)
        resultados_C.append(np.mean(Cs) / C0)
        resultados_L.append(np.mean(Ls) / L0)
    return resultados_p, resultados_C, resultados_L

valores_p = np.logspace(-4, 0, 20)  # de 0.0001 a 1, em escala log
ps, C_norm, L_norm = analise_transicao_ws(n=300, k=6, valores_p=valores_p)

plt.figure(figsize=(7, 5))
plt.semilogx(ps, C_norm, 'o-', label="C(p) / C(0) — clustering")
plt.semilogx(ps, L_norm, 's-', label="L(p) / L(0) — caminho médio")
plt.xlabel("Probabilidade de reconexão (p) — escala log")
plt.ylabel("Valor normalizado (base: p=0)")
plt.title("Transição para o regime de mundo pequeno (Watts-Strogatz)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

# Modelo de Anexação Uniforme:


```python
def gerar_anexacao_uniforme(G_inicial, n_novos_nos, m, seed=None):
    """
    Anexação uniforme: cada novo nó escolhe m nós existentes ao acaso,
    com probabilidade igual para todos (independente do grau deles).
    """
    rng = np.random.default_rng(seed)
    G = G_inicial.copy()
    proximo_id = max(G.nodes()) + 1

    for _ in range(n_novos_nos):
        novo_no = proximo_id
        nos_existentes = list(G.nodes())
        escolhidos = rng.choice(nos_existentes, size=m, replace=False)
        G.add_node(novo_no)
        for alvo in escolhidos:
            G.add_edge(novo_no, alvo)
        proximo_id += 1

    return G

UNI_1 = gerar_anexacao_uniforme(G0, n_novos_nos=500, m=3, seed=42)
grau_uni1 = [g for no, g in UNI_1.degree()]

print("Nós finais:", UNI_1.number_of_nodes())
print("Arestas finais:", UNI_1.number_of_edges())
print("Grau médio:", np.mean(grau_uni1))
print("Grau máximo:", max(grau_uni1))
```

# Modelo de _Barabási-Albert_ (Anexação Preferencial):

```python
def gerar_anexacao_preferencial(G_inicial, n_novos_nos, m, seed=None):
    """
    Anexação preferencial: a probabilidade de um nó existente ser
    escolhido é proporcional ao seu grau atual (quem tem mais conexões,
    tem mais chance de ganhar novas conexões).
    """
    rng = np.random.default_rng(seed)
    G = G_inicial.copy()
    proximo_id = max(G.nodes()) + 1

    for _ in range(n_novos_nos):
        novo_no = proximo_id
        nos_existentes = list(G.nodes())
        graus = np.array([G.degree(n) for n in nos_existentes], dtype=float)
        probabilidades = graus / graus.sum()  

        escolhidos = rng.choice(nos_existentes, size=m, replace=False, p=probabilidades)
        G.add_node(novo_no)
        for alvo in escolhidos:
            G.add_edge(novo_no, alvo)
        proximo_id += 1

    return G

BA_1 = gerar_anexacao_preferencial(G0, n_novos_nos=500, m=3, seed=42)
grau_ba1 = [g for no, g in BA_1.degree()]

print("Nós finais:", BA_1.number_of_nodes())
print("Arestas finais:", BA_1.number_of_edges())
print("Grau médio:", np.mean(grau_ba1))
print("Grau máximo:", max(grau_ba1))
```

# Modelo de Price (Anexação + Aleatória):

```python
def gerar_modelo_price(G_inicial, n_novos_nos, m, alpha, seed=None):
    """
    Modelo de Price: fração alpha das m ligações de cada novo nó segue
    anexação preferencial (proporcional ao grau); a fração restante
    (1-alpha) segue anexação uniforme (chance igual para todos).
    """
    rng = np.random.default_rng(seed)
    G = G_inicial.copy()
    proximo_id = max(G.nodes()) + 1

    for _ in range(n_novos_nos):
        novo_no = proximo_id
        nos_existentes = list(G.nodes())
        G.add_node(novo_no)

        m_preferencial = int(round(m * alpha))
        m_uniforme = m - m_preferencial

        escolhidos = set()

        if m_preferencial > 0:
            graus = np.array([G.degree(n) for n in nos_existentes], dtype=float)
            probs = graus / graus.sum()
            escolhidos_pref = rng.choice(nos_existentes, size=m_preferencial,
                                          replace=False, p=probs)
            escolhidos.update(escolhidos_pref)

        if m_uniforme > 0:
            candidatos = [n for n in nos_existentes if n not in escolhidos]
            escolhidos_unif = rng.choice(candidatos, size=m_uniforme, replace=False)
            escolhidos.update(escolhidos_unif)

        for alvo in escolhidos:
            G.add_edge(novo_no, alvo)
        proximo_id += 1

    return G

PRICE_1 = gerar_modelo_price(G0, n_novos_nos=500, m=3, alpha=0.7, seed=42)
grau_price1 = [g for no, g in PRICE_1.degree()]

print("Nós finais:", PRICE_1.number_of_nodes())
print("Grau médio:", np.mean(grau_price1))
print("Grau máximo:", max(grau_price1))
```
